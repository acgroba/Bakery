# Preparación de los datos

```python
print(df.isnull().sum().sort_values(ascending=False))
```

```python
print(df[df['Item']=='NONE'])
```

```python
df.drop(df[df['Item']=='NONE'].index, inplace=True)
```

```python
# Year
df['Year'] = df['Date'].apply(lambda x: x.split("-")[0])
# Month
df['Month'] = df['Date'].apply(lambda x: x.split("-")[1])
# Day
df['Day'] = df['Date'].apply(lambda x: x.split("-")[2])
# Day of the week (0 Monday, 6 Sunday)
df['Weekday'] = df['Date'].apply(lambda x: dt.datetime (int(x.split("-")[0]),int(x.split("-")[1]),int(x.split("-")[2])).weekday())

print(df.head())
```

# Modelado de los datos

### Primera mitad de la mañana de días de la semana

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules, apriori
transaction_list1 = []

# For loop to create a list of the unique transactions throughout the dataset:
for i in df_first_morning_week['Transaction'].unique():
    tlist = list(set(df_first_morning_week[df_first_morning_week['Transaction']==i]['Item']))
    if len(tlist)>0:
        transaction_list1.append(tlist)

te = TransactionEncoder()
te_ary = te.fit(transaction_list1).transform(transaction_list1)
df2 = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets = apriori(df2, min_support=0.01, use_colnames=True)
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1.0)
rules.sort_values('support', ascending=False)
```

```python
transaction_list2 = []

# For loop to create a list of the unique transactions throughout the dataset:
for i in df_first_morning_weekend['Transaction'].unique():
    tlist = list(set(df_first_morning_weekend[df_first_morning_weekend['Transaction']==i]['Item']))
    if len(tlist)>0:
        transaction_list2.append(tlist)

te = TransactionEncoder()
te_ary = te.fit(transaction_list2).transform(transaction_list2)
df2 = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets = apriori(df2, min_support=0.01, use_colnames=True)
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1.0)
rules.sort_values('support', ascending=False)
```
